```python
#import packages
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
from sklearn.preprocessing import LabelEncoder
from scipy import stats
from scipy.special import boxcox1p
import warnings
# Global Settings
# To ignore warnings
warnings.filterwarnings("ignore") 
#Limiting float output to 3 decimal points
pd.set_option('display.float_format', lambda x: '{:.3f}'.format(x))
# Default figure size in inches
rcParams['figure.figsize'] = 12,6 
```

```python
# Read data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
#check the numbers of samples and features
print("The train data size before dropping Id feature is : {} ".format(train.shape))
print("The test data size before dropping Id feature is : {} ".format(test.shape))

#Save the 'Id' column
train_ID = train['ID']
test_ID = test['ID']

#Now drop the  'Id' colum since it's unnecessary for  the prediction process.
train.drop("ID", axis = 1, inplace = True)
test.drop("ID", axis = 1, inplace = True)

#check again the data size after dropping the 'Id' variable
print("\nThe train data size after dropping Id feature is : {} ".format(train.shape)) 
print("The test data size after dropping Id feature is : {} ".format(test.shape))
```

```python
# Get the list of column that have null values
na_val = data.columns[data.isnull().any()]

# Get missing value count in each column
data_na_cnt = data[na_val].isnull().sum()

# Get missing values percentage for each column
data_na = (data[na_val].isnull().sum()/len(data)) * 100
data_na = pd.DataFrame({'Total Null Val': data_na_cnt, 'Null Value %': data_na})
data_na = data_na.sort_values(by='Null Value %', ascending=False)
data_na
```

```python
#display the null values
sns.barplot(x=data_na.index, y='Null Value %', data=data_na)
plt.xticks(rotation = 90)
plt.title('Column name with null value %')
plt.show()
```

```python
# Converting non-numeric predictors stored as numbers into string
#make some columns to be categorical 
for col in ('BuildingClass', 'YearSold', 'MonthSold','Quality','OverallRating'):
    data[col] = data[col].astype(str)
```

```python
# Filling Categorical NaN (That we know how to fill due to the description file )

data['Functional'] = data['Functional'].fillna('Typ')
data['Electrical'] = data['Electrical'].fillna("SBrkr")
data['KitchenQuality'] = data['KitchenQuality'].fillna("TA")
data['Exterior1'] = data['Exterior1'].fillna(data['Exterior1'].mode()[0])
data['Exterior2'] = data['Exterior2'].fillna(data['Exterior2'].mode()[0])
data['SaleType'] = data['SaleType'].fillna(data['SaleType'].mode()[0])
data["PoolQuality"] = data["PoolQuality"].fillna("None")
data["TypeOfAlleyAccess"] = data["TypeOfAlleyAccess"].fillna("None")
data['FireplaceQuality'] = data['FireplaceQuality'].fillna("None")
data['FenceQuality'] = data['FenceQuality'].fillna("None")
data['Misc Feature'] = data['Misc Feature'].fillna("None")
data['MasVnrType'] = data['MasVnrType'].fillna("None")
for col in ('GarageArea', 'GarageCars'):
    data[col] = data[col].fillna(0)
        
for col in ['GarageType', 'GarageFinish', 'GarageQuality', 'GarageCondition']:
    data[col] = data[col].fillna('None')
    
for col in ('BsmtExposure', 'BsmtCondition', 'BsmtHeight', 'BsmtFinishType1', 'BsmtFinishType2'):
    data[col] = data[col].fillna('None')
    
for col in ('BsmtFinish1Area', 'BsmtFinish2Area', 'BsmtFullBaths', 'BsmtHalfBaths', 'MasVnrArea','BsmtUnfinishedArea', 'TotalBsmtArea'):
    data[col] = data[col].fillna(0)

data['LotFrontage'] = data['LotFrontage'].fillna(train['LotFrontage'].median())
```

```python
# Checking the features with NaN remained out

for col in data:
    if data[col].isna().sum() > 0:
        print(data[col][1])
```

```python
#create new variables
data["SqFtPerRoom"] = data["LivingArea"] / (data["TotalRooms"] +data["FullBaths"] + data["HalfBaths"] +data["Kitchens"])

data['Total_Home_Quality'] = data['Quality'] + data['OverallRating']

data['Total_Bathrooms'] = (data['FullBaths'] + (0.5 * data['HalfBaths']) +
                               data['BsmtFullBaths'] + (0.5 * data['BsmtHalfBaths']))

data["HighQualityArea"] = data["1stFloorArea"] + data["2ndFloorArea"]
data['YearBiultAndRemove']= data['YearBuilt'] + data['YearRemodelled']
```

```python
# Removing the useless variables
useless = ['GarageYearBuilt','YearRemodelled'] 
data = data.drop(useless, axis = 1)
```

```python
# Creating dummy variables from categorical features

data_dummy = pd.get_dummies(data)
from scipy.stats import skew
numeric_features = data_dummy.dtypes[data_dummy.dtypes != object].index
skewed_features = data_dummy[numeric_features].apply(lambda x: skew(x)).sort_values(ascending=False)
high_skew = skewed_features[skewed_features > 0.5]
skew_index = high_skew.index
skew_index
```

```python
# Normalize skewed features using log_transformation
    
for i in skew_index:
    data_dummy[i] = np.log1p(data_dummy[i] )
```

```python
#Recheck nan 
na_val = data.columns[data.isnull().any()]

# Get missing value count in each column
data_na_cnt = data[na_val].isnull().sum()

# Get missing values percentage for each column
data_na = (data[na_val].isnull().sum()/len(data)) * 100
data_na = pd.DataFrame({'Total Null Val': data_na_cnt, 'Null Value %': data_na})
data_na = data_na.sort_values(by='Null Value %', ascending=False)
data_na
```

```python
import shap
from xgboost import XGBRegressor
from catboost import Pool
from sklearn.svm import SVR
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeRegressor
from mlxtend.regressor import StackingRegressor
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error
```

```python

#split train and test data
y_train = target_log
x_train = data_dummy[:len(y_train)]
x_test = data_dummy[len(y_train):]
print('Shape of train: {}, test:{}'.format(x_train.shape,x_test.shape))
```

```python
# Creation of the RMSE metric:
    
def rmse(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))

def cv_rmse(model):
    rmse = np.sqrt(-cross_val_score(model, x_train, y_train, scoring="neg_mean_squared_error", cv=kf))
    return (rmse)
def score(model_reg):
    score_reg = cv_rmse(model_reg)
    print('score mean: {}, std: {}'.format(score_reg.mean(),score_reg.std()))
    cv_scores.append(score_reg.mean())
    cv_std.append(score_reg.std())
```

```python
# 10 Fold Cross validation

kf = KFold(n_splits=11, random_state=42, shuffle=True)

cv_scores = []
cv_std = []

baseline_models = ['Linear_Reg.','Bayesian_Ridge_Reg.','LGBM_Reg.','SVR',
                   'Dec_Tree_Reg.','Random_Forest_Reg.', 'XGB_Reg.',
                   'Grad_Boost_Reg.','Cat_Boost_Reg.','Stacked_Reg.','Stacked_Reg2']
```

```python
# Linear Regression
linear_reg = LinearRegression()
score(linear_reg)
```

```python
# Bayesian Ridge Regression
bay_ridge_reg = BayesianRidge(compute_score=True)
score(bay_ridge_reg)
```

```python
# Light Gradient Boost Regressor
l_gbm = LGBMRegressor(objective='regression')
score(l_gbm)
```

```python
# Support Vector Regression

svr = SVR()
score(svr)
```

```python
# Decision Tree Regressor

dtr = DecisionTreeRegressor()
score(dtr)
```

```python
# Random Forest Regressor

rfr = RandomForestRegressor()
score(rfr)
```

```python
# XGB Regressor

xgb = XGBRegressor()
score(xgb)
```

```python
# Gradient Boost Regressor

gbr = GradientBoostingRegressor()
score(gbr)
```

```python
# Cat Boost Regressor

catb = CatBoostRegressor()
score(catb)
```

```python
# Stacked Regressor

stack_gen = StackingRegressor(regressors=(CatBoostRegressor(),
                                          BayesianRidge()),
                              meta_regressor = CatBoostRegressor(),
                              use_features_in_secondary = True)
score(stack_gen)
```

```python
# Stacked Regressor

stack_gen2 = StackingRegressor(regressors=(CatBoostRegressor(),
                                          XGBRegressor()),
                              meta_regressor = CatBoostRegressor(),
                              use_features_in_secondary = True)

score(stack_gen2)
```

```python
final_cv_score = pd.DataFrame(baseline_models, columns = ['Regressors'])
final_cv_score['RMSE_mean'] = cv_scores
final_cv_score['RMSE_std'] = cv_std
```
